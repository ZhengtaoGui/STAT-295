```python
import numpy as np
import prompt_engnerring as pe
import matplotlib.pyplot as plt

num_tasks = 200
feature_dim = 2
num_workers = 50
budget = 1000
budget1 = 1000
budget2 = 1000
budget3 = 1000
select_size = 10
times = 50
total_task_weight_errors = np.zeros(int(budget/select_size))
total_worker_weight_errors = np.zeros(int(budget/select_size))
total_random_errors = np.zeros(int(budget/select_size))
total_mean_errors = np.zeros(int(budget/select_size))
total_model_errors = np.zeros(int(budget/select_size))
total_task_weight_errors1 = np.zeros(int(budget/select_size))
total_worker_weight_errors1 = np.zeros(int(budget/select_size))
total_random_errors1 = np.zeros(int(budget/select_size))
total_mean_errors1 = np.zeros(int(budget/select_size))
total_model_errors1 = np.zeros(int(budget/select_size))
total_task_weight_errors2 = np.zeros(int(budget/select_size))
total_worker_weight_errors2 = np.zeros(int(budget/select_size))
total_random_errors2 = np.zeros(int(budget/select_size))
total_mean_errors2 = np.zeros(int(budget/select_size))
total_model_errors2 = np.zeros(int(budget/select_size))

for i in range(times):
    X, true_scores, task_budgets, worker_budgets, w,  beta, worker_scores = pe.generate_matching_data(
num_tasks, feature_dim, num_workers)
    task_budgets1 = task_budgets.copy()
    worker_budgets1 = worker_budgets.copy()
    worker_scores1 = worker_scores.copy()
    task_budgets2 = task_budgets.copy()
    worker_budgets2 = worker_budgets.copy()
    worker_scores2 = worker_scores.copy()
    task_weight_errors, worker_weight_errors, random_errors, mean_errors, task_weights, worker_weights, model_errors = pe.train_worker_task_matching(0.005, w, beta, worker_scores, true_scores, X, task_budgets, worker_budgets, 1000, select_size)
    task_weight_errors1, worker_weight_errors1, random_errors1, mean_errors1, task_weights1, worker_weights1, model_errors1 = pe.train_worker_task_matching_explore_all(0.1,w, beta, worker_scores1, true_scores, X, task_budgets1, worker_budgets1, 1000, select_size)
    task_weight_errors2, worker_weight_errors2, random_errors2, mean_errors2, task_weights2, worker_weights2, model_errors2 = pe.train_worker_task_matching_true_score(0.1,w, beta, worker_scores2, true_scores, X, task_budgets2, worker_budgets2, 1000, select_size)
    total_task_weight_errors += task_weight_errors
    total_worker_weight_errors += worker_weight_errors
    total_random_errors += random_errors
    total_mean_errors += mean_errors
    total_model_errors += model_errors
    total_task_weight_errors1 += task_weight_errors1
    total_worker_weight_errors1 += worker_weight_errors1
    total_random_errors1 += random_errors1
    total_mean_errors1 += mean_errors1
    total_model_errors1 += model_errors1
    total_task_weight_errors2 += task_weight_errors2
    total_worker_weight_errors2 += worker_weight_errors2
    total_random_errors2 += random_errors2
    total_mean_errors2 += mean_errors2
    total_model_errors2 += model_errors2

total_task_weight_errors = total_task_weight_errors/times
total_worker_weight_errors = total_worker_weight_errors/times
total_random_errors = total_random_errors/times
total_mean_errors = total_mean_errors/times
total_model_errors = total_model_errors/times
total_task_weight_errors1 = total_task_weight_errors1/times
total_worker_weight_errors1 = total_worker_weight_errors1/times
total_random_errors1 = total_random_errors1/times
total_mean_errors1 = total_mean_errors1/times
total_model_errors1 = total_model_errors1/times
total_task_weight_errors2 = total_task_weight_errors2/times
total_worker_weight_errors2 = total_worker_weight_errors2/times
total_random_errors2 = total_random_errors2/times
total_mean_errors2 = total_mean_errors2/times
total_model_errors2 = total_model_errors2/times

def plot_cumulative_average1(data1,data2,data3):

    # Plotting the cumulative averages
    plt.figure(figsize=(10, 6))
    plt.plot(range(0, len(data1) ), data1, marker='o', linestyle='-', label='active_select_errors_0.05LR')
    plt.plot(range(0, len(data2) ), data2, marker='o', linestyle='-', label='active_select_errors_0.15LR')
    plt.plot(range(0, len(data3) ), data3, marker='o', linestyle='-', label='random_select_errors_0.25LR')
    plt.xlabel('Number of Budget')
    plt.ylabel('Mean_Error')
    plt.title('Comparisions with active learning and random selection')
    plt.grid()
    plt.legend(loc='upper right')
    plt.show()

def plot_cumulative_average2(data1,data2,data3):

    # Plotting the cumulative averages
    plt.figure(figsize=(10, 6))
    plt.plot(range(0, len(data1) ), data1, marker='o', linestyle='-', label='task_weight_errors_0.05LR')
    plt.plot(range(0, len(data2) ), data2, marker='o', linestyle='-', label='task_weight_errors_0.15LR')
    plt.plot(range(0, len(data3) ), data3, marker='o', linestyle='-', label='task_weight_errors_0.25LR')
    plt.xlabel('Number of Budget')
    plt.ylabel('Mean_Error')
    plt.title('Comparisions about task_weight')
    plt.grid()
    plt.legend(loc='upper right')
    plt.show()

def plot_cumulative_average3(data1,data2,data3):

    # Plotting the cumulative averages
    plt.figure(figsize=(10, 6))
    plt.plot(range(0, len(data1) ), data1, marker='o', linestyle='-', label='worker_weight_errors_0.05LR')
    plt.plot(range(0, len(data2) ), data2, marker='o', linestyle='-', label='worker_weight_errors_0.15LR')
    plt.plot(range(0, len(data3) ), data3, marker='o', linestyle='-', label='worker_weight_errors_0.25LR')
    plt.xlabel('Number of Budget')
    plt.ylabel('Mean_Error')
    plt.title('Comparisions about worker_weight')
    plt.grid()
    plt.legend(loc='upper right')
    plt.show()

def plot_cumulative_average4(data1,data2,data3):

    # Plotting the cumulative averages
    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 1000, len(data1)), data1, marker='o', linestyle='-', label='model_prediction_errors_under_AL_process')
    plt.plot(np.linspace(0, 1000, len(data1)), data2, marker='o', linestyle='-', label='model_prediction_errors_without_AL_process')
    plt.plot(np.linspace(0, 1000, len(data1)), data3, marker='o', linestyle='-', label='model_prediction_errors_given_known_true_scores')
    plt.xlabel('Number of Budget')
    plt.ylabel('Mean of model_prediction_errors')
    plt.title('Comparisions of model prediction errors under AL process & without AL & given known true scores')
    plt.grid()
    plt.legend(loc='upper right')
    plt.show()


plot_cumulative_average1(total_mean_errors, total_mean_errors1, total_mean_errors2)
plot_cumulative_average2(total_task_weight_errors,total_task_weight_errors1,total_task_weight_errors2)
plot_cumulative_average3(total_worker_weight_errors,total_worker_weight_errors1,total_worker_weight_errors2)
plot_cumulative_average4(total_model_errors,total_model_errors1,total_model_errors2)
```
